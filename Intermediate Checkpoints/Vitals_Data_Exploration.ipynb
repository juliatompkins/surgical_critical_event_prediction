```python
data = pd.read_csv('/Users/dylanmack/Library/CloudStorage/OneDrive-WashingtonUniversityinSt.Louis/ESE 527/EMR/patient_vitals.csv')
data
```

```python
data_new = data[['PID']].copy()
data_new.loc[:,'Obs_time'] = (pd.to_datetime(data.loc[:,'Obs_time']))
data_new.loc[:, 'time'] = data_new.groupby('PID')['Obs_time'].transform(lambda x: x - x.min())
data_new.loc[:, 'HRe'] = data.loc[:,'HRe'].apply(lambda x: float(x) if x != '\\N' else 0)
data_new.loc[:, 'HRp'] = data.loc[:,'HRp'].apply(lambda x: float(x) if x != '\\N' else 0)
data_new.loc[:, 'nSBP'] = data.loc[:,'nSBP'].apply(lambda x: float(x) if x != '\\N' else 0)
data_new.loc[:, 'nMAP'] = data.loc[:,'nMAP'].apply(lambda x: float(x) if x != '\\N' else 0)
data_new.loc[:, 'nDBP'] = data.loc[:,'nDBP'].apply(lambda x: float(x) if x != '\\N' else 0)
data_new.loc[:, 'SP02'] = data.loc[:,'SP02'].apply(lambda x: float(x) if x != '\\N' else 0)

data_new.set_index('Obs_time', inplace=True)

print(f'There are {len(np.unique(data_new["PID"]))} patients')
```

```python
def pull_patient_data(PID):
    return data_new[data_new["PID"] == PID]
```

```python
# Begin by analyzing data of patient who experienced an adverse event (CPR)

patient = 'aa028abbb7d98a45'
patient_data = pull_patient_data(patient)

#(patient_data['Obs_time'])

for measure in ['HRe', 'HRp', 'nSBP', 'nMAP', 'nDBP', 'SP02']:
    ts = np.array(patient_data[measure])
    plt.figure()
    plt.title(measure)
    plt.plot(patient_data.index, ts)
    plt.plot(pd.to_datetime('2017-04-14 16:19:00'), patient_data[patient_data.index == '2017-04-14 16:19:00'][measure], marker = 'x', markersize = 10, color = 'red')
    plt.xticks(rotation=45)  
    
```
